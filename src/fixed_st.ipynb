```python
from pathlib import Path
import pickle
from marker_analyser.classes import OscillationCollection
from marker_analyser.configuration import FitConfig, ParamConfig, FitSegment
```

```python
results_dir = Path("\\uosfstore.shef.ac.uk\Shared\newton_lab\Processed\Metallodrugs\CuDda\20260128\Stretch_modulus")
```

```python
# Load the data
with open(
    "//uosfstore.shef.ac.uk/Shared/newton_lab/Scripts/Metallodrugs/CuDda/marker-analyser/marina_data_loaded_oscillations.pkl",
    "rb",
) as file:
    oscillations: OscillationCollection = pickle.load(file)
    assert type(oscillations) == OscillationCollection

# get only the first 10 oscillations for testing
first_10_oscillation_ids = list(oscillations.oscillations.keys())[:10]
oscillations = OscillationCollection(
    oscillations={osc_id: oscillations.oscillations[osc_id] for osc_id in first_10_oscillation_ids}
)

print(f"Loaded {len(oscillations.oscillations)} oscillations for testing.")
```

```python
# Want to fix the stretch modulus, to 1100 pN.
oscillations_fixed_st = oscillations.__deepcopy__()
fit_config_st_fixed = FitConfig(
    params_config={"St": ParamConfig(global_param=True, fixed=True, initial_value=1100)},
    auto_calculate_and_fix_f_offset=True,
    model_name="fixed st",
    segment="both",
)

oscillations_fixed_st.fit_global_model_to_all(fit_config=fit_config_st_fixed)
oscillations_fixed_st.global_fit_save_parameters_to_csv_file(results_dir / "global_fit_parameters_fixed_st.csv")
```

```python
# now don't fix the stretch modulus
oscillations_st_free = oscillations.__deepcopy__()

fit_config_st_free = FitConfig(
    params_config={"St": ParamConfig(global_param=True, fixed=False)},
    auto_calculate_and_fix_f_offset=True,
    model_name="free st",
    segment=FitSegment.BOTH,
)

oscillations_st_free.fit_global_model_to_all(fit_config=fit_config_st_free)

oscillations_st_free.global_fit_save_parameters_to_csv_file(results_dir / "global_fit_parameters_free_st.csv")
```
